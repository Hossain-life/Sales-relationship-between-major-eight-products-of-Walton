Code for drawing heatmap. I used pearson method.

```python
ax = plt.axes()
dataplot = sns.heatmap(flightsnew.corr("pearson"),cmap="YlGnBu",annot=True,fmt='.2f')
ax.set_title('Heatmap')
plt.show()
```
